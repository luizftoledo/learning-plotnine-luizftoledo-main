```python
#manual or automatic new column
mtcars['transmission_type'] = mtcars['am']\
    .apply(lambda x: {0: 'automatic', 
                      1: 'manual'}[x])

mtcars.head()
pd.set_option('display.max_rows', 7893848377)
```

```python
#TRANSMISSION TYPE X FUEL EFICIENCY

(
    ggplot(mtcars,          
        aes('mpg', 'transmission_type')) 
        + geom_point()    
        + labs(
            title = "Manual cars tend to be more fuel efficient",
            y = "Transmission type",
            x = "Fuel eficiency (Miles per gallon)"
        )
)
```

```python
#TRANSMISSION TYPE X ACCELERATION

(
    ggplot(mtcars,          
        aes('qsec', 'transmission_type')) 
        + geom_point()    
        + labs(
            title = "Manual cars tend to be slower than automatic",
            y = "Transmission type",
            x = "Acceleration performance (1/4 mile time)"
        )
)
```

```python
# The average eficiency for automatic cars is 17.14 miles per gallon, while it is 24.39 for the manual ones.
mtcars.groupby(by='transmission_type').median('mpg')
```

```python
# 7 out of 10 best fuel-efficiency car models are manual
mtcars.sort_values(by='mpg', ascending=False).head(10)
```

```python
# 8 out of 10 worst fuel-efficiency car models are automatic
mtcars.sort_values(by='mpg', ascending=False).tail(10)
```

```python
# 6 out of 10 slowest acceleration car models are automatic
mtcars.sort_values(by='qsec', ascending=False).head(10)
```

```python
# 6 out of 10 best acceleration car models are manual
mtcars.sort_values(by='qsec', ascending=False).tail(10)
```

```python
#Most automatic cars are heavier than manuals

(
    ggplot(mtcars,          
        aes('wt', 'transmission_type')) 
        + geom_point()    
        + labs(
            title = "Automatic cars are heavier",
            y = "Transmission type",
            x = "Weight"
        )
)
```

```python
# The average weight for automatic cars is 3.7 lb/1000, while manual is 2.4
mtcars.groupby(by='transmission_type').mean()
```

```python
# Weight x speed
(
    ggplot(mtcars,                   # data
        aes('wt', 'qsec'))          # aesthetics
        + geom_point(alpha = 0.3, size = 8)               # geometry
        + theme_bw()
        + labs(
            title = "There's no meaninful difference in speed related to a car's weight",
            y = "1/4 mile time",
            x = "Weight (lb/1000)"
        
        )
)   
```

```python
(
    ggplot(mtcars,          
        aes('qsec', 'mpg')) 
        + geom_point(aes(color = 'transmission_type'), size = 3)  
        + labs(
            title = "Automatic x manual cars fuel-eficiency and acceleration performance",
            y = "Transmission type",
            x = "Acceleration performance (1/4 mile time)"
        )
)
```
